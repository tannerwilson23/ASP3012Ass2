```python
import matplotlib.pyplot as plt
import numpy
```

```python
G = 6.67*10**-8
M = 10**11*2*10**33
Rd = 1.54*10**22
vo = G*M/Rd/10**10
rspace = np.arange(0.01,100,0.01)
```

```python
def c(R):
    return (0.767*vo*(0.44*(R/5)**1.3)/(1+0.235*(R/5)**2.3))**(1/2)

def e(R):
    return (vo*5*(-np.exp(-R/5)*(R/5+1)+1)/R)**(1/2)

def cn(a,v,R):
    return v*(4*np.pi*a/R*(np.log(1+R/a)-((R/a)/(R/a+1))))**(1/2)
```

```python
plt.plot(rspace,(cn(70,110,rspace)**2+c(rspace)**2)**(1/2), label = 'tot')
plt.plot(rspace,cn(70,110,rspace), label = 'Halo Contribution')
plt.plot(rspace,c(rspace), label = 'Visible Light Contribution')
plt.ylabel(r'$v_{circ} \ (km/s)$')
plt.xlabel(r'Distance from center of Galaxy (kpc)')
plt.legend(loc = 'best')
plt.savefig('4c1.pdf')
```

```python
plt.plot(rspace,(cn(70,110,rspace)**2+c(rspace)**2)**(1/2), label = 'Total Rotation Curve')
plt.xlim(0,20)
plt.plot(rspace,cn(70,110,rspace), label = 'Halo Contribution')
plt.plot(rspace,c(rspace), label = 'Visible Light Contribution')
plt.ylabel(r'$v_{circ} \ (km/s)$')
plt.xlabel(r'Distance from center of Galaxy (kpc)')
plt.legend(loc = 'best')
plt.savefig('4c2.pdf')
```
